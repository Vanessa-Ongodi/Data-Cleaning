# Basics of Data Cleaning

Dirty data can create problems in data analysis and has to be taken care of before proceeding to the next phase of handling data. This data consists of incorrect information. Sometimes in large datasets, completely removing dirty data from a source can be impossible. To make good data predictions and create better visualizations with data, the raw data acquired must be cleaned. Unclean data can impact data results data science and data analysis. Data cleaning is a part of the data preprocessing stage. The replace function is the most common way to replace wrong values in data cleaning.
To do this we will create a dataframe with the different types of errors, identify them and fix them.

```python
import numpy as np
import pandas as pd
```

## Create a dataframe of random people in 2022

```python
#data entered by random people in 2022
df = pd.DataFrame({
    'Sex': ['F', 'F', 'M', 'M', 'M', 'M', 'D'],
    'Age': [15, 35, 23, 320, 32, 19, 28],
    'Name': ['Jane', 'Ada', 'Mark', 'John', 'John', 'Peter', 'Hope'],
    'Birth_Year': [2007, 1887, 1999, 1990, 1990, 2003, 1994],
    'Interests': ['Hiking', 'Travel', 'Sports', 'Books', 'Books', 'Swimming', 'Black'],
    'Favorite_Color': ['Blue', 'Purple', 'Green', 'White', 'White', np.nan, 'Movies']
})
df
```

### Bogus Data

This is the type of data you encounter in a variable that is not genuine or not true. The variable common and the values are expected to follow a particular range. This data can happen due to a typing error. We can correct this data by replacing it with the actual value we estimate it to be or if it is private data, we can ask the owner to confirm the error.

```python
df['Age'].unique()
```

```python
df.loc[df['Age'] > 100, 'Age'] = df.loc[df['Age'] > 100, 'Age'] / 10
df
```

### Duplicate Data

As the name implies, this data is encountered when data is entered twice. This can be solved by checking for duplicates in the dataframe and removing the duplicate values.


```python
df.duplicated()
```

```python
df = df.drop_duplicates()
df
```

### Outdated Data

This type of data happens when we expect data from a particular period and we find values that are below the particular periods. We can check if the data was caused by a typing error by estimating if it falls in the range when a value is replaced. But if the value is beyond the range, we can remove it from the dataset.

```python
df['Birth_Year'].unique()
```

```python
df.loc[df['Birth_Year'] == 1887, 'Birth_Year'] = df.loc[df['Birth_Year'] == 1887, 'Birth_Year'] + 100
df
```

### Incorrect Data

Incorrect data is data entered in the wrong column For example, when numerical data is entered in a text column or vice versa or when a value that should be in another column is entered in a different column.

```python
df
```

```python
df = df.replace({
    'Interests': {
        'Black': 'Movies'
    },
    'Favorite_Color': {
        'Movies': 'Black'
    }
})
```

```python
df
```

### Inconsistent Data

This data is encountered when the data entered assumes the shape of a data but is not the same. This data is usually caused by a typing error.

```python

df['Sex'].unique()
```

```python
df = df.replace({
    'Sex': {
        'D': 'F'
    }})
```

```python
df
```

### Missing Data

This is a type of data that is returned as a null value. We can find the null values in the dataframe, remove them or replace them with an arbitrary value.

```python
df.isnull()
```

```python
df.fillna('Empty')
```
